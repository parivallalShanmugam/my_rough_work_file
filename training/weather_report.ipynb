```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:/python practice/data file/weather.csv")
```

```python
df.info()
```

```python


# Load the weather dataset


# Display basic information and summary
print(df.info())
print(df.describe())

# Checking for missing values
missing_values = df.isnull().sum()
print(f'Missing values in each column:\n{missing_values}')

# Unique values in 'Weather'
unique_weather = df['Weather'].unique()
print(f'Unique weather conditions:\n{unique_weather}')

# Frequency of different weather conditions
weather_counts = df['Weather'].value_counts()
print(f'Frequency of weather conditions:\n{weather_counts}')

# Mean, median, mode, standard deviation, and variance of key columns
mean_visibility = df['Visibility_km'].mean()
median_visibility = df['Visibility_km'].median()
mode_visibility = df['Visibility_km'].mode()[0]
std_visibility = df['Visibility_km'].std()
variance_visibility = df['Visibility_km'].var()

print(f'Mean Visibility: {mean_visibility}')
print(f'Median Visibility: {median_visibility}')
print(f'Mode Visibility: {mode_visibility}')
print(f'Standard Deviation of Visibility: {std_visibility}')
print(f'Variance of Visibility: {variance_visibility}')

# Visualize the distribution of Visibility
plt.figure(figsize=(10, 6))
sns.histplot(df['Visibility_km'], bins=30, kde=True)
plt.title('Distribution of Visibility')
plt.xlabel('Visibility (km)')
plt.ylabel('Frequency')
plt.show()

# Scatter plot to explore relationship between Temperature and Visibility
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temp_C', y='Visibility_km', data=df)
plt.title('Temperature vs Visibility')
plt.xlabel('Temperature (C)')
plt.ylabel('Visibility (km)')
plt.show()





```

```python
# Time series analysis for Visibility
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df.set_index('Date/Time', inplace=True)
df['Visibility_km'].plot(figsize=(15, 6))
plt.title('Visibility Over Time')
plt.xlabel('Date')
plt.ylabel('Visibility (km)')
plt.show()
```

```python
# Analyzing outliers in Wind Speed
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Wind Speed_km/h'])
plt.title('Boxplot of Wind Speed')
plt.xlabel('Wind Speed (km/h)')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```
