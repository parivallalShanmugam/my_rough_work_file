```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load the CSV file
#file_path = mnt/data'//BigBasket Products.xlsx - BigBasket Products.csv'
df = pd.read_csv('D:/python practice/BigBasket Products.csv')

# Calculate mean and standard deviation of the sale_price and rating columns
mean_sale_price = df['sale_price'].mean()
std_dev_sale_price = df['sale_price'].std()

mean_rating = df['rating'].mean()
std_dev_rating = df['rating'].std()

# Define the range for the x-axis for both columns
x_sale_price = np.linspace(min(df['sale_price']),max(df['sale_price']), 100)
x_rating = np.linspace(min(df['rating']),max(df['rating']), 100)

# Plot the normal distribution for sale_price
plt.plot(x_sale_price, stats.norm.pdf(x_sale_price, mean_sale_price, std_dev_sale_price), label='Sale Price Normal Distribution')

# Plot the normal distribution for rating
plt.plot(x_rating, stats.norm.pdf(x_rating, mean_rating, std_dev_rating), label='Rating Normal Distribution')

# Plot the histogram of the sale_price data
plt.hist(df['sale_price'], bins=30, density=True, alpha=0.6, color='g', label='Sale Price Data')

# Plot the histogram of the rating data
plt.hist(df['rating'], bins=30, density=True, alpha=0.6, color='b', label='Rating Data')

# Add title and labels
plt.title('Normal Distribution of Sale Price and Rating')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()

# Print mean and standard deviation
print("Sale Price - Mean:", mean_sale_price, "Standard Deviation:", std_dev_sale_price)
print("Rating - Mean:", mean_rating, "Standard Deviation:", std_dev_rating)
```

```python
import numpy as np
from scipy import stats

# Sample data: scores before and after an intervention
before = np.array([20, 21, 22, 23, 24, 25])
after = np.array([20, 21, 22, 23, 24, 24])

# Perform the paired t-test
t_stat, p_value = stats.ttest_rel(before, after)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpreting the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the means.")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

# Sample data
data1 = {'scores': [68, 72, 74, 65, 70, 80, 75, 78, 85, 77]}
data2 = {'scores': [60, 62, 64, 59, 66, 70, 71, 74, 73, 75]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Calculate statistics
mean1 = df1['scores'].mean()
mean2 = df2['scores'].mean()
std1 = df1['scores'].std(ddof=0)
std2 = df2['scores'].std(ddof=0)
n1 = len(df1)
n2 = len(df2)

# Compute the Z-statistic
standard_error = np.sqrt((std1**2 / n1) + (std2**2 / n2))
z_statistic = (mean1 - mean2) / standard_error

# Calculate the p-value
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

# Make a decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

# Print results
print("Mean1:", mean1)
print("Mean2:", mean2)
print("Z-Statistic:", z_statistic)
print("P-Value:", p_value)
```
