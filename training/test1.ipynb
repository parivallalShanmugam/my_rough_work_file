```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df.dtypes
```

```python
df = df.astype({'cost_price':'float','list_price':'float','discount_percent':'float'})
```

```python
y = []

for i in range(len(df)):
    x = tuple(df.iloc[i])
    y.append(x)

y    
```

```python
file = open('orderdata.txt','w')

for tuple in y:
    file.write(str(tuple)+','+'\n')

file.close()    
```
