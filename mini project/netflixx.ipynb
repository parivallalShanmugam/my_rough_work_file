```python
import pandas as pd
import mysql.connector
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
y = []

for i in range(len(df)):
    x = tuple(df.iloc[i])
    y.append(x)

y    
```

```python
file = open('netflix.txt','w')

for tuple in y:
    file.write(str(tuple)+','+'\n')

file.close() 
```

```python
with open('netflix.txt', 'w', encoding='utf-8') as file:
    for item in y:
        file.write(str(item) + ',\n')
```
